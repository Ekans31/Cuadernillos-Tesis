# Invariante de Kretschmann

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap

def f(r, mu):
    A = (
        -1085 * r**6
        - (441/2) * r**2 * mu
        + (3/2) * np.sqrt(7) * r**2 * np.sqrt(
            532656 * r**8 + 30380 * r**4 * mu + 3087 * mu**2
        )
    )
    
    B = (
        -2170 * r**6
        - 441 * r**2 * mu
        + 3 * np.sqrt(7) * r**2 * np.sqrt(
            532656 * r**8 + 30380 * r**4 * mu + 3087 * mu**2
        )
    )
    
    return (
        1
        + r**2
        + A**(1/3) / (7**(2/3))
        - 101 * (2/7)**(1/3) * r**4 / (B**(1/3))
    )

def Krt(r,mu):
    fr = f(r,mu)
    df = np.gradient(fr, r) 
    d2f = np.gradient(df, r)
    
    # Krt(r)
    K = (12*(fr - 1)**2 + 6*(r**2)*(df**2) + (r**4)*(d2f**2)) / (r**4)
    return K
```

```python
r_vals = np.linspace(0, 3, 600)
mu_list = [0.05, 0.1, 0.2, 0.5, 1, 2, 4, 8,16]

plt.rcParams.update({
    "font.family": "serif",
    "font.size": 12,
    "figure.dpi": 300,
    "axes.labelsize": 14,
    "axes.titlesize": 16,
    "legend.fontsize": 10,
    "lines.linewidth": 2
})

# Colormap profesional tipo arcoiris
cmap = get_cmap("turbo")   # O "viridis", "plasma", "rainbow"

# Graficar
plt.figure(figsize=(7,5))

for i, mu in enumerate(mu_list):
    color = cmap(i / (len(mu_list) - 1))
    plt.plot(r_vals, f(r_vals, mu), color=color, label=f"$\\mu = {mu}$")

plt.xlabel("$r$")
plt.ylabel("$f(r)$")
plt.title("Funcin $f(r)$ para varios valores de $\\mu$")
plt.grid(True, alpha=0.3)
plt.legend(title="Valores de $\\mu$", frameon=True)
plt.tight_layout()

plt.savefig("grafico_f_mu.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
r_vals = np.linspace(0, 16, 10000)
mu_list = [0.05, 0.1, 0.2, 0.5, 1, 2, 4, 8,16]

plt.figure(figsize=(7,5))

for i, mu in enumerate(mu_list):
    color = cmap(i / (len(mu_list) - 1))
    plt.plot(np.sqrt(r_vals), np.log(Krt(r_vals, mu)), color=color, label=f"$\\mu' = {mu}$")

plt.xlabel("$Sqrt(r')$")
plt.ylabel("$Log(Krt(r'))$")
plt.title("Invariante de Kretschmann para varios valores de $\\mu'$")
plt.grid(True, alpha=0.3)
plt.legend(title="Valores de $\\mu'$", frameon=True)
plt.tight_layout()

plt.savefig("grafico_Krt_mu.png", dpi=300, bbox_inches="tight")
plt.show()
```
