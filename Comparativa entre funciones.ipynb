# Comparaci칩n entre funciones originadas con distintas densidades de tercer orden con acoplamiento cuadr치tico

```python
import numpy as np
import matplotlib.pyplot as plt

# Par치metro
mu = 2.0   # puedes cambiarlo

# Dominio
r = np.linspace(0.01, 5, 5000)  # evitamos r=0 por potencias fraccionarias

# ----------- f3rees1 -----------
def f3rees1(r, mu):
    A = -1085*r**6 - (441*r**2*mu)/2 \
        + (3/2)*np.sqrt(7)*r**2*np.sqrt(
            532656*r**8 + 30380*r**4*mu + 3087*mu**2
        )

    B = -2170*r**6 - 441*r**2*mu \
        + 3*np.sqrt(7)*r**2*np.sqrt(
            532656*r**8 + 30380*r**4*mu + 3087*mu**2
        )

    return 1 + r**2 \
           + (A)**(1/3)/7**(2/3) \
           - (101*(2/7)**(1/3)*r**4)/(B**(1/3))


# ----------- f3rees2 -----------
def f3rees2(r, mu):
    C = -2219*r**4 - 2268*mu \
        + 18*np.sqrt(14)*np.sqrt(
            15174*r**8 + 2219*r**4*mu + 1134*mu**2
        )

    return 1 + r**2 \
           - (209*r**(10/3))/(7**(1/3)*C**(1/3)) \
           + (r**(2/3)*C**(1/3))/7**(2/3)


# Evaluaci칩n
f1 = f3rees1(r, mu)
f2 = f3rees2(r, mu)
```

```python
plt.figure(figsize=(6,5))

plt.plot(r, f1, linewidth=2.0, color="#1f3b73",
         label=r'$Z_{3}$')

plt.plot(r, f2, linewidth=2.0, linestyle='--',
         color="#8b1e3f",
         label=r"$Z'_{3}$")

# L칤nea horizontal en y = 0
plt.axhline(0, color='gray', linestyle='--', linewidth=1.2)

plt.xlabel(r'$r$', fontsize=13)
plt.ylabel(r'$f(r)$', fontsize=13)

plt.grid(True, linestyle=':', linewidth=0.6, alpha=0.7)
plt.legend(frameon=False, fontsize=11)

plt.tick_params(direction='in')
plt.tight_layout()
plt.show()
```
